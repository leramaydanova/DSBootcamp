# Ex05. Оптимизация Pandas

## 1. Чтение файла fines.csv

```python
fines_df = pd.read_csv("data/fines.csv", header=0, sep=',')
```

## 2. Итерации с вычислениями по строкам фрейма данных

```python
def func(row_):
    return row_['Fines'] / row_['Refund'] * row_['Year']
```

### 2.1. Цикл

```python
new_column = []
for i in range(0, len(fines_df)):
    math_value = func(fines_df.iloc[i])
    new_column.append(math_value)

fines_df.assign(Math=new_column)
```

### 2.2. Цикл с методом iterrows()

```python
new_column = []
for index, row_ in fines_df.iterrows():
    math_value = func(row_)
    new_column.append(math_value)
fines_df.assign(Math=new_column)
```

### 2.3. Apply(lambda x: ())

```python
new_column = fines_df.apply(lambda x: func(x), axis=1)
fines_df.assign(Math=new_column)
```

### 2.4. Вычисления с помощью Series из фрейма данных

```python
new_column = fines_df['Fines'] / fines_df['Refund'] * fines_df['Year']
fines_df.assign(Math=new_column)
```

### 2.5. Вычисления с помощью свойства values фрейма данных

```python
new_column = [row_[0] / row_[1] * row_[2] for row_ in fines_df[['Fines', 'Refund', 'Year']].values]
fines_df = fines_df.assign(Math=new_column)
fines_df
```

## 4. Downcasting типов данных в фрейме данных

### 4.1. Изначальные dtypes и использование памяти

```python
fines_df.info(memory_usage='deep')
```

### 4.2. Понижение float64 до float32 и int64 до минимально возможного

```python
optimized_df = fines_df.copy()
float64_cols = optimized_df.select_dtypes(include='float64').columns
optimized_df[float64_cols] = optimized_df[float64_cols].astype('float32')

int64_cols = optimized_df.select_dtypes(include='int64').columns
optimized_df[int64_cols] = optimized_df[int64_cols].apply(lambda x: pd.to_numeric(x, downcast='integer'))
```

### 4.3. Dtypes и использование памяти после оптимизации float и integer

```python
optimized_df.info(memory_usage='deep')
```

### 4.4. Оптимизация object данных

```python
object_cols = optimized_df.select_dtypes(include='object').columns
optimized_df[object_cols] = optimized_df[object_cols].astype('category')
```

### 4.5. Dtypes и использование памяти после оптимизации object

```python
optimized_df.info(memory_usage='deep')
```
