# Ex05. Диаграмма размаха

## 1. Чтение csv файла

```python
ab_test = pd.read_csv('../data/ab-test.csv', header=0, sep=',')
```

## 2. Диаграмма размаха для тестовой и контрольной групп

**Стили для диаграммы**

```python
boxplot_style = {
    'whiskerprops': {'color': 'red'},
    'medianprops': {'linewidth': 2, 'color': 'red'},
    'boxprops': {'linewidth': 3, 'facecolor': (0, 0, 0, 0), 'edgecolor': 'green'},
    'capprops': {'color': 'red'}
}
```

```python
ab_test.groupby('time').boxplot(by='group', figsize=(15, 8), fontsize=10, 
    patch_artist=True,
    showmeans=True,
    **boxplot_style)
plt.suptitle('Lag before labs deadlines. After and Before for Newsfeed', fontsize=15);
```
