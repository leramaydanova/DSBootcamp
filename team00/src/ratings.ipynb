```python
from ratings import Ratings, Math
```

```python
ratings_path = '../ml-latest-small/ratings.csv'
movies_path = '../ml-latest-small/movies.csv'
count_rows = 1000
movies_analyzer = Ratings.Movies(ratings_path, movies_path, count_rows)
users_analyzer = Ratings.Users(ratings_path, movies_path, count_rows)
```

```python
%%timeit

movies_analyzer = Ratings.Movies(ratings_path, movies_path, count_rows)
users_analyzer = Ratings.Users(ratings_path, movies_path, count_rows)
```

# Movies



**Kоличество выпущенных в прокат фильмов по годам**

```python
top = movies_analyzer.dist_by_year(100)
for item in top.items(): 
    print(f"{item[0]} год выпуска: фильмов {item[1]}")
```

```python
%%timeit

top_n = 100
top = movies_analyzer.dist_by_year(top_n)
```

**Рейтинги по количеству фильмов, которым выставлена данная оценка**

```python
top_n = 100
top = movies_analyzer.dist_by_rating(top_n)
for item in top.items(): 
    print(f"Оценка {item[0]}: фильмов {item[1]}")
```

```python
%%timeit

top_n = 100
top = movies_analyzer.dist_by_rating(top_n)
```

**Топ фильмов по количеству оценок**

```python
top_n = 10
top = movies_analyzer.top_by_num_of_ratings(top_n)
print(f"Топ {top_n}:\n")
for item in top.items(): 
    print(f"{item[0]}: оценок {item[1]}")
```

```python
%%timeit

top_n = 10
top = movies_analyzer.top_by_num_of_ratings(top_n)
```

**Топ фильмов по среднему рейтингу**

```python
top_n = 10
top = movies_analyzer.top_by_ratings(top_n)
print(f"Топ {top_n}:\n")
for item in top.items(): 
    print(f"{item[0]}: средний рейтинг {item[1]}")
```

```python
%%timeit

top_n = 10
top = movies_analyzer.top_by_ratings(top_n)
```

**Топ фильмов по медианному рейтингу**

```python
top_n = 10
top = movies_analyzer.top_by_ratings(top_n, Math.median)
print(f"Топ {top_n}:\n")
for item in top.items(): 
    print(f"{item[0]}: медианный рейтинг {item[1]}")
```

```python
%%timeit

top_n = 10
top = movies_analyzer.top_by_ratings(top_n, Math.median)
```

**Топ фильмов по значению дисперсии всех оценок**

```python
top_n = 10
top = movies_analyzer.top_controversial(top_n)
print(f"Топ {top_n}:\n")
for item in top.items(): 
    print(f"{item[0]}: дисперсия {item[1]}")
```

```python
%%timeit

top_n = 10
top = movies_analyzer.top_controversial(top_n)
```

**Топ фильмов с наибольшим диапозоном проставленных оценок**

```python
top_n = 10
top = movies_analyzer.top_range(top_n)
print(f"Топ {top_n}:\n")
for item in top.items(): 
    print(f"{item[0]}: разница в диапозоне {item[1]}")
```

```python
%%timeit

top_n = 10
top = movies_analyzer.top_range(top_n)
```
